# List Comprehension

## Case 1: Creating a new list that incrementing 1 from passed argument

```python
# Normal List method
numbers  = [1, 3, 5]
new_list = []
for x in numbers:
    new_list.append(x + 1)
new_list
```

```python
# list comprehension
new_compr = [x + 1 for x in numbers]
print(new_compr)
```

List comprehension Explanation:


`new_list` : the variable to store the result <br>
`new_item` : the new result (method) after being processed (<font color = violet>**x+1**</font>)<br>
`item` : the object iteration from the looping processes (<font color = violet>**x**</font>) <br>
`list`: the argument that used for processing the result (<font color = violet>**numbers**</font>)

## Case 2: List comprehension on the string

```python
lst_cmp = [x for x in "Hazeya"]
lst_cmp
```

Explanation:
- Returning the result as iteration from passed argument
- Since the argument is "string", the list will be splitting each letter with coma into the list

## Case 3: doubling the number in the passed range using list comprehension

```python
print([x*2 for x in range (1, 5)])
```

Explanation:
- `for x in range (1, 5)`: Iterating each number inside 1-4 range
- `x * 2`: Iterated number dikali 2
- Langsung print

## Case 1: Taking the first three letters from passed arguments

```python
full_stats = ["ATTACK", "DEFENSE", 'INTELLIGENCE', "DURABILITY", "AGILITY", "DEXTERITY", "HP", "LUCK"]
stat = [x[:3] for x in full_stats if len(x) > 3]
stat
```

Explanation:
- `for x in full_stats`: looping the value from full stat 
- `if len(x) > 3`: collecting the value from full stat, when the text length containing >3 letters
- `x[:3]`: taking the first 3 letters when previous condition fulfilled
- will be excluding any value which has less than 3 text lengths ("HP")

## Case 2: Converting the first three letters from passed arguments into a uppercase

```python
full_stats = ['attack', 'defense', 'intelligence', 'agility', 'dexterity', 'hp', 'luck']
stat = [x[:3].upper() for x in full_stats if len(x) > 2]
stat
```

Explanation:
- `for x in full_stats` looping the value from full stat
- `if len(x) > 2` collecting the value from full stat, when the text length containing >2 letters  
- `x[:3].upper()` taking the first 3 letters when previous condition fulfilled then convert it into the titlecase 
- will be excluding any value which has less than 2 text lengths ("HP")
