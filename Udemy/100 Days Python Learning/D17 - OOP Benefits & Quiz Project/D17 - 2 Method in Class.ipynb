# Methods
The attributes are the things that the object has and the <font color = "lime">***methods***</font> are the things that the object does.
Methods: Function that operate on the attributes of a specific instance of a class



## Methods Format
This method looks exactly the same as how we've created other functions. We have the Def keyword, we have the name of the method, we have some parentheses which can take inputs, a colon, and then the body of the method. So in this case, all it does is it gets hold of the object and then the object seats attribute


```python
class User:
    def __init__(self, user_id, name):
        self.id = user_id
        self.name = name
        self.followers = 0  # default attributes without passing argument
        self.following = 0  # attributes ini akan nempel ke mana-mana
        
    def follow(self, user):  # passing another instance as the argument
        user.followers += 1  # another instance value is increasing
        self.following += 1  # self instance is increasing
        
    def stat(self):
        return f"{self.name} following: {self.following} people, and having {self.followers} followers"
```

```python
user_1 = User("001", "Peter")
user_2 = User("002", "Parker")
```

```python
user_1.follow(user_2)
```

```python
print(user_1.stat())
print(user_2.stat())
```
