# Class
Class bisa disebut juga sebagai blueprint untuk membuat temporary object. Sesi ini membahas untuk membuat blueprint sendiri sehingga kita bisa membuat object yang kita inginkan



````python
class User:
    pass
````

pass dimaksudkan untuk memberikan command kosong, karena akan teerjadi error jika kita memberikan function / class seperti di atas tanpa memberikan command berikutnya [*"error indentation"*]

````python
user_1 = User()
user_1.id = "001"
user_1.name = "Me"
````

# Constructor
is a part of the blueprint that allows us to specify what should happen when our object is being constructed. And this is also known in programming as initializing an object. When the object is being initialized, we can set variables or counters to they're starting values. In Python, the way that we would create the constructor is by using a special function, which is the <font color = "magenta">***init***</font> function.



````python
class User:
    def __init__(self):
        print("New user is being created...")
````

````python
user_1 = User()
user_1.id = "001"
user_1.name = "Tom"
print(user_1.name)
````

````python
user_2 = User()
user_2.id = "002"
user_2.name = "Morello"
print(user_2.name)
````

Dalam case diatas attributes (printing) dalam Constructor akan selalu terbawa jika class diassign ke dalam variable baru

# Attributes in the Constructor
We have our init function and inside the init function in addition to this thing called <font color = "lime">***self***</font> which is the actual object that's being created or being initialized,

In addition, you can add as many parameters as you wish. And that parameter is going to be passed in when an object gets constructed from this class. And once you receive that data, then you can use it to set the object's attributes


Pada contoh di atas, kita meng-assign langsung jumlah seats ke dalam Car attributes, untuk writing assignment nya, bisa dibuat seperti ini:

```yaml
class Car:
    def __init__(self, seats):
        self.seats = seats

my_car = Car()
my_car = Car (5) "atau" my_car.seats = 5
```

## Contoh pembuatan Constructor & assignment nya

````python
class User:
    def __init__(self, user_id, name):
        self.id = user_id
        self.name = name
````

````python
user_1 = User("001", "Tommy")
````

````python
user_1.id
````

````python
user_1.name
````

## Contoh Constructor Assignment yg salah

````python
user_2 = User()
user_2.id = "002"
user_2.name = "Hanks"
````

Hal diatas terjadi karena:
- constructor sudah disusun menggunakan 2 attributes
- karena menggunakan 2 attributes, kita juga harus memberikan 2 instances pada variable yang kita gunakan
