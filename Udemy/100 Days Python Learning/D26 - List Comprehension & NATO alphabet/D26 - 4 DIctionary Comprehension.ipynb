# Dictionary Comprehension

## Simple Format Writing

 <br>
- `new_dict` : the variable to store the result 
- `new_key` : the new key data after being processed 
- `new_value`: the new value data after being processed 
- `item` : the object iteration (key only) from the  "for looping" processes
- `list`: the argument (variable) that used for processing the result

### Case 1: Generating random points from the given stats

```python
import random as ran

full_stats = ["ATTACK", "DEFENSE", 'INTELLIGENCE', "AGILITY", "DEXTERITY", "LUCK"]
char1 = {x : ran.randint(0,99) for x in full_stats}
char1
```

Explanation:
- `for x in full_stats` : looping the contents from passed variable
- `x :` : collectiong the iterated variable content as a new key
- `: ran.randint(0,99)` : adding new random values between 0-98 from iterated key for it's pair
- `char1` : variable to store the new dictionary

## Advance Format Writing

 <br>
- `new_dict` : the variable to store the result 
- `new_key` : the new key data after being processed 
- `new_value`: the new value data after being processed 
- `(key,value)` : the objects iteration (key & value) from the "for looping" processes
- `dict.items()`: the argument (value) that used for processing the result

## Advance Format Writing with conditional statement

 <br>
- `new_dict` : the variable to store the result 
- `new_key` : the new key data after being processed 
- `new_value`: the new value data after being processed 
- `(key,value)` : the objects iteration (key & value) from the "for looping" processes
- `dict.items()`: the argument (value) that used for processing the result
- `test`: additional condition <font color = "cyan">***"which only can be used"***</font> to proccessing the result (new_dict & key)

### Case 2: Collecting passed dictionary values which reaching specific conditions

```python
char_overall = {x : y for (x, y) in char1.items() if y >= 50}
char_overall
```

Explanation:
- `for (x, y) in char1.items()` : Iterating the key (x) and value (y) in the passed dictionary
- `char1.items() if y >= 50` : Collecting any dictionary values that more than equal 50
- `x : y`: storing any key and values which passed previous condition
- `char_overall` : variable that used to store a new proccesed contents from the executed script <br>
`.items()` is used because we want to collect the values from the dictionary. Without it, we will receiving error `too many values to unpack`
