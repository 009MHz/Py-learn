# Modules Import

## 1. Basic Import
Basic Import is only using keyword and module, then calling it by assign a variable that contains the modules name & the Class name from the modules


```python
import turtle

tim = turtle.Turtle()
```

## 2. from *{module}* import *{module_content}*
Similar with previous import but on this method we inserting directly the contents (class) so we don't need to assign twice on the variable when we want to using it


```python
from turtle import turtle

leo = Turtle()
rafa = Turtle()
donnie = Turtle()
```

## 3. Importing whole contents from called module `from module import *`
Same with previous method, but the different thing is, we will inserting all availables contents from the called modules. And we can use everything inside the modules


```python
from turtle import *
from random import *

forward(100)
choice([1,2,3,5])
```

- The cons dari method ini adalah, "ketidak jelasan fungsi yang dipakai". Code diatas bisa berfungsi normal, tapi akan sulit mengidentifikasi function apakah yg dimaksud code diatas jika sebuah fitur memiliki >90 lines tanpa menggunakan class yg dipakai untuk menggunakan fungsi tersebut.
- `forward` merupakan method dari module turtle
- `choice` merupakan method dari module random

## 4. Aliasing modules
This method is pretty similar with method one, but we will using aliases as the replacement for the called module to shorten the variable / values that we will use later


```python
import turtle as t

don = t.Turtle()
```

This method is suitable to use on the long name modules such as `pyscreenshot, VideoCapture` etc and made our script is ***eye-friendly***
