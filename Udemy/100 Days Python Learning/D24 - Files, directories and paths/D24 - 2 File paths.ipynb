# File Paths
is the direction in the computer to reach specific items (folder/file)

### Absolute File Path
File path ilustrations:
- `root` folder : `/`
- Work folder: `/Work`
- report files: `/Work/report.doc`
- Project folder: `/Work/Project`
- talk file: `/Work/Project/talk.ppt`


### Relative file paths
Ilustrations:
- The file path is shorten by only using filenames only `./talk.ppt`
- The rest of the path is shorten because we're currently inside the working directory (project folder)
